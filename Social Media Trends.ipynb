```python
df.rename(columns={"UsageDuraiton": "UsageDuration"}, inplace=True)
```

```python
# Distribution of Usage Duration & Total Likes

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["UsageDuration"], bins=10, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Distribution of Usage Duration")

sns.histplot(df["TotalLikes"], bins=10, kde=True, ax=axes[1], color="orange")
axes[1].set_title("Distribution of Total Likes")

plt.tight_layout()
plt.show()
```

```python
# Age vs. Usage Duration
sns.scatterplot(x=df["Age"], y=df["UsageDuration"], color="green")
plt.title("Age vs Usage Duration")
plt.xlabel("Age")
plt.ylabel("Usage Duration (hours)")
plt.show()
```

```python
# Total Likes per Country

country_likes = df.groupby("Country")["TotalLikes"].sum().sort_values(ascending=False)
sns.barplot(x=country_likes.values, y=country_likes.index, palette="coolwarm")

plt.title("Total Likes per Country")
plt.xlabel("Total Likes")
plt.ylabel("Country")
plt.show()
```

```python
# Identifying Outliers in Usage Duration & Total Likes

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(y=df["UsageDuration"], ax=axes[0], color="blue")
axes[0].set_title("Box Plot: Usage Duration")

sns.boxplot(y=df["TotalLikes"], ax=axes[1], color="orange")
axes[1].set_title("Box Plot: Total Likes")

plt.tight_layout()
plt.show()
```
